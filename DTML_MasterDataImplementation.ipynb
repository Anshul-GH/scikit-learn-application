# Eliminating issues with Skew and Scaling

<b>Eliminating Skew via log conversion</b>

```python
# the code converts the numeric features values to log equivalents and also takes care of the negative values
X = X.apply(lambda x: ((1+x)/(1+abs(x)))*(np.log(1 + abs(x))))
```

```python
X.head(20)
```

<b>Eliminating Scaling issues using RobustScaler</b>

```python
# Again, scaling only applies to numerical data.
from sklearn.preprocessing import RobustScaler

# initializing the scaler object
scaler = RobustScaler()

# define a new dataframe to hold the scaled numeric features
X_scaled = pd.DataFrame(data = X)

# apply the scaling and populate the data
X_scaled = scaler.fit_transform(X)
```
