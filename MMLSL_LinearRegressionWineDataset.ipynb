```python
# fitting and evaluating the model
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_csv('/home/anshul/MMLSL/chapter05/winequality-red.csv', sep=';')
```

```python
# df.head()
X = df.drop(['quality'], axis=1) # axis=1 implies 'columns'
y = df['quality']
```

```python
# split the dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y) # split into default 75/25 set with a random state.

regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print('R-Squared: ', regressor.score(X_test, y_test))
```

```python
# implementing cross-validation to better judge the model performance
from sklearn.model_selection import cross_val_score

scores = cross_val_score(regressor, X, y, cv=5)
print('Scores: ', scores)
print('Mean Score: ', scores.mean())
```

```python
# print(len(y_pred))
j = 0
for i in y_test:
    print('Predicted: ', y_pred[j], '\t', 'Actual: ', i)
    j += 1
```

```python
# implementing Stochastic Gradient Descent using SGDRegressor from 'linear_model'
from sklearn.linear_model import SGDRegressor
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston

data = load_boston()
```

```python
# split the dataset into train/test
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target)
```

```python
X_scaler = StandardScaler()
y_scaler = StandardScaler()

X_train = X_scaler.fit_transform(X_train)
y_train = y_scaler.fit_transform(y_train.reshape(-1, 1))
# y_train = column_or_1d(y_train, warn=True)

X_test = X_scaler.fit_transform(X_test)
y_test = y_scaler.fit_transform(y_test.reshape(-1, 1))
# y_test = column_or_1d(y_test, warn=True)

regressor = SGDRegressor(loss='squared_loss', max_iter=5, tol=None)

scores = cross_val_score(regressor, X_train, y_train, cv=5)

print('Cross Validation R-Squared Scores: ', scores)
print('Mean Cross Validation R-Squared Score: ', scores.mean())

regressor.fit(X_train, y_train)

print('Test Set R-Squared Scores: ', regressor.score(X_test, y_test))
```
